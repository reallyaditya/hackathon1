```python
import pandas as pd

# Define the first dictionary
dict1 = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Gender": ["F", "M", "M"],
}

# Define the second dictionary
dict2 = {
    "Name": ["Dave", "Eve", "Frank"],
    "Age": [40, 45, 50],
    "Gender": ["M", "F", "M"],
}

# Merge the dictionaries into a dataframe
df = pd.merge(pd.DataFrame(dict1), pd.DataFrame(dict2), how="outer")

# Print the merged dataframe
print(df)
```

```python
pd.concat([pd.DataFrame(dict1), pd.DataFrame(dict2)])
```

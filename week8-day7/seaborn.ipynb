```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'])
plt.xlabel('Total Bill')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill with Custom Bins and Color')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True)
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.distplot(data['total_bill'], color='green')
plt.title('Univariate Distribution of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='hist')
plt.title('Bivariate Distribution of Total Bill and Tips', pad=70)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
titanic = sns.load_dataset('titanic')

# Creating a bar plot
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Class')
plt.show()
```

```python
titanic
```

```python
titanic.groupby('class')['age'].median().to_frame()
```

```python
first_class = titanic[titanic['class'] == 'First'][['age']]
```

```python
first_class
```

```python
median_age = first_class.median().values[0]
median_age
```

```python
# Creating a box plot
sns.boxplot(x='class', y='age', data=titanic)
plt.annotate(f'Median Age = {median_age}', xy=(0, median_age), xytext=(1, 29), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title('Age Distribution by Class')
plt.show()
```

```python
# Creating a violin plot
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading example dataset
data = sns.load_dataset('tips')

# Creating the scatter plot
sns.lmplot(x='total_bill', y='tip', data=data)

# Adding plot title and labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Displaying the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading example dataset
data = sns.load_dataset('tips')

# Creating the joint plot
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')

# Adding a title
plt.subplots_adjust(top=0.9)
plt.suptitle('Joint Plot of Total Bill and Tip')

# Displaying the plot
plt.show()
```

## Customizing Point Colors and Sizes

```python
import numpy as np

# Sample data
np.random.seed(0)
x = np.random.rand(25)
y = np.random.rand(25)
colors = np.random.rand(25)
sizes = 1000 * np.random.rand(25)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()  # Shows color scale
plt.title("Customized Scatter Plot")
plt.show()
```

## Adding Trend Lines and Annotations

```python
# Generating a linear trend line
z = np.polyfit(x, y, 1)  # Linear fit
p = np.poly1d(z)
plt.plot(x, p(x), "r--")  # Red dashed trend line
# Adding annotations
plt.scatter(x, y)
```

```python
# Generating a linear trend line
z = np.polyfit(x, y, 1)  # Linear fit
p = np.poly1d(z)
plt.plot(x, p(x), "r--")  # Red dashed trend line

# Adding annotations
plt.scatter(x, y)
plt.annotate('Special Point', (x[10], y[10]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("Scatter Plot with Trend Line and Annotations")
plt.show()
```

# Pie Charts with Matplotlib

## Creating a Basic Pie Chart

```python
import matplotlib.pyplot as plt

# Data
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [40, 30, 15, 10, 5]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title('Smartphone Market Share 2023')
plt.show()
```

```python
explode = (0.1, 0, 0, 0, 0)  # Only explode the first slice (Apple)

plt.pie(market_share, labels=brands, autopct='%1.1f%%', explode=explode)
plt.title('Smartphone Market Share 2023')
plt.show()
```

```python
colors = ['gold', 'lightblue', 'lightgreen', 'orange', 'purple']

plt.pie(market_share, labels=brands, autopct='%1.1f%%', colors=colors)
plt.title('Smartphone Market Share 2023')
plt.show()
```

# Box Plots in Matplotlib

## Creating Box Plots with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(100)

plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')
plt.show()
```

## Customizing Box Plots

```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

data = [data1, data2, data3]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

# Creating axes instance and the box plot
bp = ax.boxplot(data, patch_artist = True, notch = True, vert = 0)

# Changing color and line width of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='blue', linewidth = 1.5, linestyle =":")

# Changing color and line width of the caps
for cap in bp['caps']:
    cap.set(color ='red', linewidth = 2)

# Changing color and linewidth of the medians
for median in bp['medians']:
    median.set(color ='green', linewidth = 3)

# Changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)

# Adding title 
plt.title('Customized Box Plot')
plt.show()
```

# Understanding Subplots and Multiple Axes

## Basic Subplots

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# First subplot
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('First Subplot')

# Second subplot
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Second Subplot')

# Show the figure
plt.show()
```

## Advanced Layouts with Subplots

```python
# Creating a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plotting different types of plots
axs[0, 0].plot([1, 2, 3], [3, 2, 1])
axs[0, 1].bar(['A', 'B', 'C'], [3, 2, 1])
axs[1, 0].scatter([1, 2, 3], [1, 4, 9])
axs[1, 1].hist([1, 1, 2, 3, 3, 3, 4, 4], bins=[1, 2, 3, 4, 5])

plt.show()
```

## Sharing Axes

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)

ax1.plot([1, 2, 3], [2, 3, 4])
ax2.plot([1, 2, 3], [4, 3, 2])

plt.show()
```

## Adjusting Layout

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [3, 2, 1])
axs[0, 1].bar(['A', 'B', 'C'], [3, 2, 1])
plt.tight_layout(pad=0.1)
plt.show()
```

## Combining Different Types of Plots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Line plot
ax1.plot([1, 2, 3], [3, 2, 1])
ax1.set_title('Line Plot')

# Bar plot
ax2.bar(['A', 'B', 'C'], [3, 2, 1])
ax2.set_title('Bar Chart')

plt.show()
```

## Interactive and Animated Plots in Matplotlib

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Initial data
t = np.linspace(0, 1, 1000)
initial_freq = 5
y = np.sin(2 * np.pi * initial_freq * t)

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)
l, = ax.plot(t, y)

# Slider
axfreq = plt.axes([0.1, 0.1, 0.65, 0.03])
freq_slider = Slider(axfreq, 'Freq', 0.1, 30.0, valinit=initial_freq)

# Update function
def update(val):
    freq = freq_slider.val
    l.set_ydata(np.sin(2 * np.pi * freq * t))
    fig.canvas.draw_idle()

# Connect the update function to the slider
freq_slider.on_changed(update)

plt.show()
```

## Animated Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)

plt.show()
```

```python
import pandas as pd
import kaggle
```

```python
kaggle.api.dataset_download_files('osmi/mental-health-in-tech-survey', path='.', unzip=True)
```

```python
df = pd.read_csv("survey.csv")
```

```python
df.info()
```

```python
df = df[(df['Age'] > 0) & (df['Age'] < 100)]
```

```python
import matplotlib.pyplot as plt
df['Age'].plot(kind='hist', bins=20)
```

```python
df.describe(include='all').T
```

```python
df.isnull().sum()
```

```python
# Handling missing values
df.ffill(axis=0, inplace=True)
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df['Gender'].unique()
```

```python
# Correcting inconsistencies in gender entries
df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].replace(dict.fromkeys(['male', 'm', 'man'], 'male'))
df['Gender'] = df['Gender'].replace(dict.fromkeys(['female', 'f', 'woman'], 'female'))
```

```python
dict.fromkeys(['female', 'f', 'woman'], 'female')
```

```python
df.columns
```

```python
# Drop irrelevant columns
df.drop(columns=['Timestamp', 'state'], inplace=True)
```

```python
# Age Group Distribution
#categorize ages into bins
age_bins = [0, 18, 25, 35, 45, 55, 65, 75, 85, 95, 105]
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins)

# Group by AgeGroup and treatment, then count and normalize
age_group_dist = df.groupby(['AgeGroup', 'treatment']).size().groupby(level=0).apply(lambda x: x / x.sum())
```

```python
age_group_dist.to_frame()
```

```python
df['Age'].plot(kind='hist', bins=age_bins, edgecolor='black')
```

```python
# Display the results
print("Age Group Distribution of Mental Health Conditions:\n")
age_group_dist
```

```python
# Gender-Based Frequency
gender_dist = df.groupby('Gender')['treatment'].value_counts(normalize=True)

print("\nGender-Based Frequency of Mental Health Issues:\n")
gender_dist.to_frame()
```

```python
# Country-Wise Analysis
country_dist = df.groupby('Country')['treatment'].value_counts().to_frame()

print("\nCountries with Highest and Lowest Reported Rates:\n")
new_country_dist = country_dist.reset_index()
country_dist.sort_values(by='count', ascending=False)
```

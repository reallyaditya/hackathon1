```python
import pandas as pd

weather_data = pd.read_csv('weatherHistory.csv')
```

```python
weather_data
```

```python
weather_data['Formatted Date'] = pd.to_datetime(weather_data['Formatted Date'])
```

```python
weather_data.describe()
```

```python
weather_data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
weather_data['Temperature_normalized'] = scaler.fit_transform(weather_data[['Temperature (C)']])
```

```python
weather_data['Temperature_normalized'].describe()
```

```python
weather_data
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
reduced_data = pca.fit_transform(weather_data[['Temperature_normalized', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)']])
```

```python
pd.DataFrame(reduced_data)
```

```python
average_temperature_by_precipitation = weather_data.groupby('Precip Type')['Temperature (C)'].mean()
average_temperature_by_precipitation.to_frame()
```

```python
average_temperature_by_precipitation.to_csv('average_temperature_by_precipitation.csv')
```
